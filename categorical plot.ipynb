```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
sns.barplot(x='sex',y='total_bill',data=tips)
```

```python
sns.barplot(x='sex',y='total_bill',data=tips,estimator=np.std)
```

```python
sns.countplot(x='sex',data=tips)
```

```python
tips['sex'].value_counts()
```

```python
sns.countplot(x='smoker',data=tips)
```

```python
tips['smoker'].value_counts()
```

```python
sns.boxplot(x="day", y="total_bill", data=tips,vert=True,palette='rainbow',hue='sex',legend=False)
plt.show();
```

```python
sns.boxplot(x='sex',y='total_bill',data=tips,vert=True,palette='coolwarm')
```

```python
sns.boxplot(x="day", y="total_bill", data=tips,orient='v',palette='rainbow')
```

```python
sns.boxplot( data=tips,orient='h',palette='rainbow')
```

```python
sns.boxplot(x='day',y='total_bill',hue='sex',data=tips,palette='cool')
```

```python
sns.boxplot(x='day',y='total_bill',hue='smoker',data=tips,palette='cool')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='cool')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='coolwarm',hue='sex')
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='cool',hue='sex',split=True)
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=False)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True)
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,jitter=True,hue='sex',palette='Set1')
```

```python
sns.stripplot(x="day", y="total_bill", data=tips,jitter=True,hue='sex',palette='Set1',dodge=True)
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill", data=tips)
```

```python
sns.swarmplot(x="day", y="total_bill",hue='sex',data=tips, palette="Set1", dodge=True)
```

```python
sns.violinplot(x='day',y='total_bill',data=tips,palette='rainbow')
sns.swarmplot(x='day',y='total_bill',data=tips,color='black',size=3)
```

```python
sns.catplot(x='day',y='total_bill',data=tips,kind='bar')
```
