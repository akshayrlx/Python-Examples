```python
import pandas as pd
import numpy as np
```

```python
sal=pd.read_csv('Salaries.csv')
sal
```

```python
avg_basepay=sal['BasePay'].mean()
print(avg_basepay)
```

```python
overtimepay=sal['OvertimePay'].max()
print(overtimepay)
```

```python
joseph=sal[sal['EmployeeName']=='JOSEPH DRISCOLL']['JobTitle']
print(f"joseph job is:{joseph}")
```

```python
joseph=sal[sal['EmployeeName']=='JOSEPH DRISCOLL']['TotalPayBenefits']
print(f"joseph makes:{joseph}")
```

```python
sal.loc[sal['TotalPayBenefits'].idxmax()]
```

```python
sal.loc[sal['TotalPayBenefits'].idxmin()]
```

```python
average_salary_by_year = sal.groupby('Year')['BasePay'].mean()
print(average_salary_by_year)
```

```python
number_of_unique_job_titles = sal['JobTitle'].nunique()
print(f"number of unique job titles:{number_of_unique_job_titles}")
```

```python
top_5_job_titles = sal['JobTitle'].value_counts().head(5)
print(top_5_job_titles)
```

```python
sal_2013 = sal[sal['Year'] == 2013]
job_title_counts = sal_2013['JobTitle'].value_counts()
job_titles_one_person = job_title_counts[job_title_counts == 1]
num_job_titles_one_person = job_titles_one_person.count()
print(num_job_titles_one_person)
```

```python
chief_titles = sal[sal['JobTitle'].str.contains('Chief', case=True, na=False)]
num_chief_titles = chief_titles.shape[0]
print(f"Number of people with 'Chief' in their job title: {num_chief_titles}")
```

```python
sal['JobTitle_length'] = sal['JobTitle'].str.len()
correlation = sal['JobTitle_length'].corr(sal['BasePay'])
print(f"Correlation between job title length and salary: {correlation}")
```
