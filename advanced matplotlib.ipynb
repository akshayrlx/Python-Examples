```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
x = np.linspace(1, 5, 11)
y = x ** 2
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18,color="blue")

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.2f$"  % y for y in yticks], fontsize=18);
```

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("scientific notation")

ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter) 
```

```python
matplotlib.rcParams['xtick.major.pad'] = 5
matplotlib.rcParams['ytick.major.pad'] = 5

fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("label and axis spacing")

ax.xaxis.labelpad = 50
ax.yaxis.labelpad = 5

ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel");
```

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("title")
ax.set_xlabel("x")
ax.set_ylabel("y")

fig.subplots_adjust(left=0.15, right=.5, bottom=0.5, top=0.6);
```

```python
fig, ax = plt.subplots(1, 2)
      
ax[0].plot(x, x**2, x, np.exp(x))
ax[0].set_yticks([0, 50, 100, 150])
ax[1].plot(x, x**2, x, np.exp(x))
ax[1].set_yticks([0, 50, 100, 150])
ax[0].set_title("title")
ax[1].set_title("title2")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

fig.subplots_adjust(left=0.15, right=.5, bottom=0.5, top=0.9);
```
