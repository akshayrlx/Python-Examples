```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

## Exercise 1

```python
fig=plt.figure()
axis=fig.add_axes([0,0,1,1])
axis.plot(x,y)
axis.set_xlabel('X')
axis.set_ylabel('Y')
axis.set_title('title')

```

```python
fig=plt.figure()
axis=fig.add_axes([0,0,1,1])
axis.plot(x,y)
axis.set_xlim([0,100])
axis.set_ylim([0,200])
axis.set_title('title')
plt.yticks(np.arange(0,max(y)+50,50))
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax1.plot(x, y)
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax2.plot(y, x)
ax2.set_xlabel("x")
ax2.set_ylabel("y")
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])

ax1.plot(x, y,'r')
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_xlim([0,100])
ax1.set_ylim([0,200])

ax1.set_yticks((np.arange(0,max(y)+50,50)))
ax2.plot(x,y,'r')
ax2.set_xlabel("x")
ax2.set_ylabel("y")

ax2.set_xlim([0,100])
ax2.set_ylim([0,200])

ax2.set_xticks((np.arange(0,max(x)+20,20)))
ax2.set_yticks((np.arange(0,max(y)+50,50)))
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

ax1.plot(x, z,'b')
ax1.set_xlabel("x")
ax1.set_ylabel("y")

ax2.plot(x,y)
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("zoom")
ax2.set_xlim(20,22)
ax2.set_ylim(30,50)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2)
plt.tight_layout()

ax[0].plot(x,y,linewidth=5,
           linestyle='-',alpha=0.1,marker='o',
           markersize=20,markerfacecolor='yellow',
           markeredgewidth=5,markeredgecolor='green')
ax[1].plot(x, z, 'v')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y,'b',lw=2.50, ls='--')
axes[0].set_xlim(0,100)
axes[0].set_ylim(0,200)

axes[1].plot(x,z,'r')
axes[1].set_xlim(0,100)
axes[1].set_ylim(0,10000)
```

```python
fig, ax = plt.subplots(nrows = 1, ncols = 2, figsize = (10, 3))
plt.tight_layout()

ax[0].plot(x,y,linewidth=5,linestyle='-',alpha=0.1,marker='o',
           markersize=20, markerfacecolor='yellow',
           markeredgewidth=5, markeredgecolor='green')
ax[1].plot(x, z, 'v')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(10,2))

axes[0].plot(x,y,'b',lw=3.50)
axes[0].set_xlim(0,100)
axes[0].set_ylim(0,200)

axes[1].plot(x,z,'r',lw=1.50, ls='--')
axes[1].set_xlim(0,100)
axes[1].set_ylim(0,10000)
plt.tight_layout()
```
