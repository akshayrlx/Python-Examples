```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
x = np.linspace(1, 5, 11)
y = x ** 2
fig, axes = plt.subplots(1, 2, figsize=(10,4))
      
axes[0].plot(x, x**2, x, np.exp(x))
axes[0].set_title("Normal scale")

axes[1].plot(x, x**2, x, np.exp(x))
axes[1].set_yscale("log")
axes[1].set_title("Logarithmic scale (y)");
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18,color="blue")

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.2f$"  % y for y in yticks], fontsize=18);
```

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("scientific notation")

ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1))
ax.yaxis.set_major_formatter(formatter) 
```

```python
matplotlib.rcParams['xtick.major.pad'] = 5
matplotlib.rcParams['ytick.major.pad'] = 5

fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("label and axis spacing")

ax.xaxis.labelpad = 50
ax.yaxis.labelpad = 5

ax.set_xlabel("xlabel")
ax.set_ylabel("ylabel");
```

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_yticks([0, 50, 100, 150])

ax.set_title("title")
ax.set_xlabel("x")
ax.set_ylabel("y")

fig.subplots_adjust(left=0.15, right=.5, bottom=0.5, top=0.9);
```

```python
fig, ax = plt.subplots(1, 2)
      
ax[0].plot(x, x**2, x, np.exp(x))
ax[0].set_yticks([0, 50, 100, 150])
ax[1].plot(x, x**2, x, np.exp(x))
ax[1].set_yticks([0, 50, 100, 150])
ax[0].set_title("title")
ax[1].set_title("title2")
ax[0].set_xlabel("x")
ax[0].set_ylabel("y")

fig.subplots_adjust(left=0.15, right=.5, bottom=0.5, top=0.9);
```

```python
fig,axes=plt.subplots(1,2,figsize=(10,3))

axes[0].plot(x,x**2,x,x**3,lw=2)
axes[0].grid(True)

axes[1].plot(x,x**2,x,x**3,lw=2)
axes[1].grid(alpha=0.5,color='g',linestyle='dashed',lw=2)
```

```python
fig,axes=plt.subplots(figsize=(5,4))

axes.spines['bottom'].set_color('blue')
axes.spines['top'].set_color('green')

axes.spines['left'].set_color('red')
axes.spines['left'].set_linewidth(2)

axes.spines['right'].set_color("none")
axes.yaxis.tick_left() 
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('red')
ax.spines['top'].set_color('red')

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))

ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

xx = np.linspace(-0.75, 1., 100)
ax.plot(xx, xx**3);
```

```python
fig, ax = plt.subplots()

ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')

xx = np.linspace(-0.75, 1., 100)
ax.plot(xx, xx**3);


ax.yaxis.set_ticks_position('left')
ax.spines['left'].set_position(('data',0))

ax.xaxis.set_ticks_position('bottom')
ax.spines['bottom'].set_position(('data',0))
```

```python
n = np.array([0,1,2,3,4,5])
```

```python
fig, axes = plt.subplots(1, 2, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))

axes[1].step(n, n**2, lw=2)
axes[1].set_title("step")
```

```python
fig, axes = plt.subplots(1, 4, figsize=(12,3))

axes[0].scatter(xx, xx + 0.25*np.random.randn(len(xx)))

axes[1].step(n, n**2, lw=2,color='#A96464')
axes[1].set_title("step")

axes[2].bar(n, n**2, align="center", width=0.5, alpha=0.5,color='#9DCF1E')
axes[2].set_title("bar")

axes[3].fill_between(x, x**2, x**4, color="#EAE317", alpha=0.5)
axes[3].set_title("fill_between")
fig.tight_layout();
```

## grid spec

```python
import matplotlib.gridspec as gridspec
```

```python
fig = plt.figure()

gs = gridspec.GridSpec(3, 4, height_ratios=[2,1,3], width_ratios=[1,5,1,2])
for g in gs:
    ax = fig.add_subplot(g)
    
fig.tight_layout()
```
